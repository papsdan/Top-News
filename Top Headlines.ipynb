```python
import requests
```

```python
from keys import API_KEY
```

```python
URL = 'https://newsapi.org/v2/top-headlines'
##example url: https://newsapi.org/v2/top-headlines?country=de&category=business&apiKey=03162c9a1db04dc29914a05d9f2b7f06
```

```python
## A function to get the top headlines by using API key
def top_headlines(country, category):
    params = {'country': country,'category': category,'apiKey':API_KEY}
    response = requests.get(URL,params)
    data = response.json()
    return data

##function accessing the json data and formatting into a list
def format_data(data):
    article_list = []
    for i in data['articles']:
        title = i['title']
        author = i['author']
        image = i['urlToImage']
        insert_dict = {
            'title':title,
            'author':author,
            'image':image
        }
        article_list.append(insert_dict)
    return article_list
```

```python
country_strings = 'ae ar at au be bg br ca ch cn co cu cz de eg fr gb gr hk hu id ie il in it jp kr lt lv ma mx my ng nl no nz ph pl pt ro rs ru sa se sg si sk th tr tw ua us ve za'
country_codes = country_strings.split(' ')
category_string = 'business entertainment general health science sports technology'
category_codes = category_string.split(' ')
```

```python
country = input('Please enter a country reference - Possible options: ae ar at au be bg br ca ch cn co cu cz de eg fr gb gr hk hu id ie il in it jp kr lt lv ma mx my ng nl no nz ph pl pt ro rs ru sa se sg si sk th tr tw ua us ve za') ##Allows user to input a country (e.g. gb, de, etc.)
while country not in country_codes:
    country = input('Country code not recognised. Please try again:')
category = input('Please enter a category - Possible options: business, entertainment, general, health, science, sports, technology') ##Allows user to input a news category (e.g. business, sports, etc)
while category not in category_codes:
    country = input('Category code not recognised. Please try again:')
data = top_headlines(country,category)
formatted_data = format_data(data)
```
